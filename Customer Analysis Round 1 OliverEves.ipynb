```python
# Correct the values in the column customer lifetime value. They are given as a percent,
# so multiply them by 100 and change dtype to numerical type

                
def clean_column_3(value):
    if type(value) == str and value.find('%') != -1:
        return 100*float(value.replace('%', ''))
    else:
        return 100*float(value)
    
df1['customer lifetime value'] = df1['customer lifetime value'].apply(clean_column_3)
df1.head()
```
